Столбец `cityCenters_nearest` назван с нарушением стиля.

Необходимо заменить тип данных:
* в столбце `first_day_exposition` с object на datetime
* в столбцах `floors_total`, `balcony`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, `days_exposition`, `airports_nearest`, `cityCenters_nearest` c float на int
* в столбце `is_apartment` с object на bool как указано в документации

Количество значений в столбцах различается. Значит, в данных есть пропущенные значения.

Чтобы двигаться дальше, нужно устранить проблемы в данных.

## Предобработка данных

### Переименование столбцов

Переименуем колонку `cityCenters_nearest` в соответствии с правилами 'змеиного' стиля.

```python
data = data.rename(columns={'cityCenters_nearest': 'city_centers_nearest'})
```

### Пропуски значений
Посчитаем, сколько в таблице пропущенных значений в процентах

```python
(data.isna().sum()/data.shape[0]*100).round(1)
```

Пропуски в данных вызваны разными причинами, можно разделить их на две больших группы. В столбцах `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, `airports_nearest`, `cityCenters_nearest` скорее всего произошла ошибка при автоматическом сборе данных. Данные в остальных столбцах были необязательными для ввода и пользователь их проигнорировал. Это подтверждается и тем, что в столбцах `last_price`, `total_area`, `first_day_exposition`, `rooms` пропусков нет, так как это основные сведения, которые невозможно не указать.

#### Срок экспозиции

Пропуски в столбце `days_exposition` заменяем на среднее по цене квартиры.

```python
cut_bins = [i for i in range(0, 30000000, 5000000)] + \
    [y for y in range(30000000, int(data['last_price'].max()), 50000000)]
exposition_transformed = data.groupby(pd.cut(data['last_price'], bins=cut_bins)
                                      )['days_exposition'].transform('median')
data['days_exposition'] = data['days_exposition'].fillna(
    exposition_transformed)
```

#### Количество балконов

Для некоторых пропущенных значений можно предположить логичную замену. Например, если человек не указал число балконов — скорее всего, их нет. Такие пропуски заменяем на 0. 

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].isna().sum()
```

#### Апартаменты

Рассмотрим столбец `is_apartment` - является ли квартира апартаментами - нежилым помещением.

```python
data['is_apartment'].unique()
```

```python
data[data['is_apartment'] == True]['is_apartment'].count()
```

В столбце `is_apartment` встречаются три типа значений - True, False и Nan. Так как доля апартаментов на рынке аренды невелика и сравнение жилых квартир и апартаментов не входит в наши задачи, отсутствующие значения можно заменить на отрицательный ответ - False.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'].isna().sum()
```

#### Название местности

```python
print('В столбце locality_name',
      data['locality_name'].isna().sum(), 'пропусков')
```

В столбце `locality_name` можно заменить пропуски только в определенной отдаленности от центра, когда квартира будет еще точно попадать в черту города. За чертой города проставить местоположение сложно и для проекта не имеет смысла, так как мы рассматриваем все факторы либо в целом для всех местностей, либо отдельно для Санкт-Петербурга.

Рассчитаем на какой минимальной удаленности от центра встречаются значения иные, чем 'Санкт-Петербург'.

```python
data[(data['locality_name'] != 'Санкт-Петербург') &
     ~data['locality_name'].isna()].sort_values('city_centers_nearest').head(10)
```

Ближе всего оказался поселок Шушары, по нему и делаем отсечку. Заменяем все пропуски ближе 11226 км на Санкт-Петербург.

```python
data.loc[data['city_centers_nearest'] < 11226.0,
         'locality_name'] = data.loc[data['city_centers_nearest'] < 11226.0,
                                     'locality_name'].fillna('Санкт-Петербург')
print('В столбце locality_name',
      data['locality_name'].isna().sum(), 'пропуска')
```

Оставшиеся пропуски заменяем на значение 'Не задано'

```python
data['locality_name'] = data['locality_name'].fillna('Не задано')
```

#### Автоматически собранная информация

Пропуски в автоматически собранной информации - удаленность от центра, аэропорта, парков и водоемов можно заполнить медианой по населенному пункту, кроме Санкт-Петербурга, где разброс будет очень большой.

```python
list_to_transform = ['city_centers_nearest', 'airports_nearest', 'parks_around3000',
                     'parks_nearest', 'ponds_around3000', 'ponds_nearest']
for value in list_to_transform:
    value_transformed = data.groupby('locality_name')[
        value].transform('median')
    data.loc[data.loc[:, 'locality_name'] != 'Санкт-Петербург',
             value] = data[value].fillna(value_transformed)
```

В столбцах `parks_around3000` и `ponds_around3000` нужно заменить 0 на 1 если в столбцах `parks_nearest` и `ponds_nearest` есть значения меньше 3000.

```python
data[data['parks_around3000'] == 0]['parks_around3000'].count()
```

```python
data.loc[(~data['parks_nearest'].isna()) & (data['parks_nearest'] < 3000) &
         (data['parks_around3000'] == 0), 'parks_around3000'] = 1
data[data['parks_around3000'] == 0]['parks_around3000'].count()
```

```python
data[data['ponds_around3000'] == 0]['ponds_around3000'].count()
```

```python
data.loc[(~data['ponds_nearest'].isna()) & (data['ponds_nearest'] < 3000) &
         (data['ponds_around3000'] == 0), 'ponds_around3000'] = 1
data[data['ponds_around3000'] == 0]['ponds_around3000'].count()
```

Все незаполненые значения в `parks_around3000` и `ponds_around3000` можно заполнить 0

```python
data.loc[:, 'ponds_around3000'] = data.loc[:, 'ponds_around3000'].fillna(0)
```

```python
data.loc[:, 'parks_around3000'] = data.loc[:, 'parks_around3000'].fillna(0)
```

#### Высота потолков

В столбце `ceiling_height` можно поставить среднюю высоту потолков с учетом отдаленности от центра. Так в центре в квартирах старого фонда потолки будут около 3х метров, в хрущевках - 2,55 метров, а в новостройках - 2,7.

Сначала уберем неправдоподобные значения, минимальная высота потолков в жилом помещении не может быть меньше 2,4 м.Стандартная средняя высота потолков в России 2,65м.

```python
std_height = 2.65
```

```python
data.loc[data['ceiling_height'] < 2.4, 'ceiling_height'] = std_height
```

```python
data.loc[data['ceiling_height'] > 10, 'ceiling_height'] = std_height
```

Разделим квартиры по удаленности от центра и на каждые 500 метров высчитаем медианную высоту потолков. Заполним пропущенные значения этой медианой.

```python
cut_bins = [i for i in range(0, int(data['city_centers_nearest'].max()), 500)]
ceiling_transformed = data.groupby(pd.cut(data['city_centers_nearest'], bins=cut_bins)
                                   )['ceiling_height'].transform('median')
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_transformed)
```

Те пропуски, что невозможно заполнить медианой, заполним стандартным средним значением.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(std_height)
```

#### Количество этажей

Количество этажей `floors_total` также можно соотнести с удаленностью от центра, так в старом фонде и в хрущевках будет обычно по 5 этажей, а вот в новостройках количество этажей может быть любое, но в любом случае количество этажей должно быть не меньше значения в столбце `floor`. Однако так как обозначение этажа как последнего будет влиять на наши выводы, считаю правильным для новостроек это значение оставлять пропущеным.

На графике зависимости этажности от расстояния от центра отсекаем значения больше 30 этажей. В Санкт-Петербурге несколько подобных зданий, но ни одно из них не совпадает по этажности с представленными, а значит это просто ошибки.

```python
data[data['locality_name'] == 'Санкт-Петербург'].plot(kind='scatter', x='floors_total',
                                                      y='city_centers_nearest', figsize=(12, 5),
                                                      grid=True, xlim=(0, 30))
```

На графике видно, что ближе 4000 км от центра домов выше 10 этажей не бывает, а дальше 4000 км не бывает домов ниже. Посчитаем медиану для количества этажей для домов ближе 4000 м от центра и поставим ее в пропуски, если указанный этаж ниже или равен медиане, в противном случае примем его за верхний.

```python
med_floor_center = data[(data['locality_name'] == 'Санкт-Петербург') &
                        (data['city_centers_nearest'] < 4000)]['floors_total'].median()
```

```python
data.loc[(data['locality_name'] == 'Санкт-Петербург') &
         (data['city_centers_nearest'] < 4000) &
         (data['floor'] <= med_floor_center), 'floors_total'] = data[(data['locality_name'] == 'Санкт-Петербург') &
                                                                     (data['city_centers_nearest'] < 4000) &
                                                                     (data['floor'] <= med_floor_center)]['floors_total'].fillna(med_floor_center)

data.loc[(data['locality_name'] == 'Санкт-Петербург') &
         (data['city_centers_nearest'] < 4000) &
         (data['floor'] > med_floor_center), 'floors_total'] = data['floors_total'].fillna(data['floor'])
```

## Расчёты и добавление результатов в таблицу

### Расчет цены квадратного метра

Напишем функцию для расчета квадратного метра

```python
def calculate_sqrmeterprice(row):
    """
    Функция возвращает стоимость квадратного метра
    """
    try:
        return int(row['last_price'] / row['total_area'])
    except:
        return
```

Добавим стоимость квадратного метра в столбец `sqrmeter_price`.

```python
data['sqrmeter_price'] = data.apply(calculate_sqrmeterprice, axis=1)
data.head()
```

### Расчет дня недели, месяца и года публикации объявления

Создаем столбцы `weekday`, `month` и `year` и помещаем в них день недели, месяц и год размещения объявления соответственно.

```python
data['weekday'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
data.head()
```

### Определение этажа

Напишем функцию для определения этажа - первый, последний или другой. В случае, если общее количество этажей в доме неизвестно и этаж не первый, определение не удастся и функция вернет None.

```python
def what_floor(row):
    if row['floor'] == 1:
        return 'первый'
    if not pd.isna(row['floors_total']):
        if row['floors_total'] == row['floor']:
            return 'последний'
        if row['floors_total'] != row['floor']:
            return 'другой'
    return
```

Создадим столбец `first_last_floor`, в который поместим категорию этажа.

```python
data['first_last_floor'] = data.apply(what_floor, axis=1)
data.head()
```

### Соотношение площадей

Посчитаем отношение жилой площади к общей.

```python
def calc_living_to_total(row):
    try:
        return row['living_area'] / row['total_area']
    except:
        return
```

```python
data['living_to_total_area'] = data.apply(calc_living_to_total, axis=1)
data.head()
```

Посчитаем отношение площади кухни к общей.

```python
def calc_kitchen_to_total(row):
    try:
        return row['kitchen_area'] / row['total_area']
    except:
        return
```

```python
data['kitchen_to_total_area'] = data.apply(calc_kitchen_to_total, axis=1)
data.head()
```
