### Эффективность обогащения

Эффективность обогащения рассчитывается по формуле

$Recovery = \frac{C * (F - T)}{F * (C - T)} * 100\%$

где:

C — доля золота в концентрате после флотации/очистки (rougher.output.concentrate_au);

F — доля золота в сырье/концентрате до флотации/очистки (rougher.input.feed_au);

T — доля золота в отвальных хвостах после флотации/очистки (rougher.output.tail_au).

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. 

```python
recovery = pd.DataFrame()
```

```python
recovery['from_data'] = data_train['rougher.output.recovery'].copy()
```

```python
recovery['my_calc'] = ((data_train['rougher.output.concentrate_au'] *
                        (data_train['rougher.input.feed_au'] - data_train['rougher.output.tail_au'])) / 
            (data_train['rougher.input.feed_au'] *
             (data_train['rougher.output.concentrate_au'] - data_train['rougher.output.tail_au']))) * 100
```

```python
recovery.head()
```

По первым строкам созданной таблицы видно, что исходные и посчитанные нами значения очень похожи. Найдем MAE между нашими расчётами и значением признака.

```python
mae = mean_absolute_error(recovery['from_data'], recovery['my_calc'])
mae
```

Метрика МАЕ имеет очень маленькую величину, значит отклонение минимально, следовательно эффективность обогащения посчитана правильно.

## Модель

### Функция sMAPE

Напишем функцию для вычисления итогового sMAPE - симметричного среднего абсолютного процентного отклонения по следующим формулам:

$sMAPE = \frac{1}{N} \sum_{n=1}^{N}\frac{|y_i - \hat{y_i}|}{(|y_i| + |\hat{y_i}|) / 2} * 100\%$

Итоговое $sMAPE = 25\% * sMAPE(rougher) + 75\% * sMAPE(final)$

```python
def sMAPE(y_target, y_pred):
    n = len(y_target)
    y_target = np.array(y_target)
    return 100/n * np.sum(np.abs(y_target - y_pred)/((np.abs(y_target) + np.abs(y_pred))/2))
```

```python
def total_sMAPE(smape_r, smape_f):
    return 0.25*smape_r + 0.75*smape_f
```

```python
sMAPE_score = make_scorer(sMAPE, greater_is_better=False)
```
